# Hangman Game

```python
import random

word_list = ["apple", "grapes", "mango"] 

stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

chosen_word = random.choice(word_list) # Choosing a random word from the list
display = [] # List of word that user will enter
    
for letter in chosen_word : # Creating display to show blank spaces = number of letters in word
        display.append('_')
    
lives = 7

print("Welcome To Hangman!!!")

while lives > 0:
    
    print(f"\nYou have {lives} life remaining.")
    guess = input("Guess a letter : ").lower() # Asking user to guess a letter
    
    for position in range(len(chosen_word)) : # Checking if the guess letter is present in chosen word
        if chosen_word[position] == guess :
            display.pop(position)
            display.insert(position, chosen_word[position])
        
    if chosen_word.find(guess) == -1 : # What to do if guess word is not present in chosen word
        print("You guess the wrong letter!")
        print(stages[lives-1])
        lives -= 1
        if lives == 0 : # Lose Condition
            print("You Lost")
    else :
        print(' '.join(display)) # Converting list into string
    
    if "_" not in display : # Win Condition
        print("\nYou Won")
        lives = 0
```
